```python
def read_pubkey_content(file_path):
    """Reads the entire content of a public key file."""
    try:
        with open(file_path, 'r') as f:
            public_key_data = f.read()
        return public_key_data
    except FileNotFoundError:
        return "Error: File not found."
    except Exception as e:
        return f"An error occurred: {e}"

# Example usage for an OpenSSH key file
key_path = "/home/ali/.ssh/id_rsa.pub"
key_content = read_pubkey_content(key_path)

print(key_content)
```
